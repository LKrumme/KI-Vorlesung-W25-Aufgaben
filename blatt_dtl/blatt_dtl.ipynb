### Gain
Keine Lust den Gain so oft zu berechnen, deswegen mache ich ein bisschen Python


````python
import math 

S_1 = ["O", "O", "M", "M", "O", "O", "M"]

def entropy(S: list):
    result = 0
    for i in set(S):
        p = S.count(i) / len(S)
        result += p*math.log(p,2)
    return -1*result
````

````python
entropy(S_1)
````

````python
def gain(S: list, A: list):
    result = 0
    for i in set(A):
        ind = [j for j, val in enumerate(A) if val==i]
        S_v = [S[j] for j in ind]
        result += (len(S_v) / len(S)) * entropy(S_v)
    return (entropy(S) - result)
````

````python
S_kontrolle = ["A", "A", "A", "B", "B", "A"]
a_x1_kontrolle = [0,1,0,1,0,0]
print(entropy(S_kontrolle))
print(gain(S_kontrolle, a_x1_kontrolle))
````
